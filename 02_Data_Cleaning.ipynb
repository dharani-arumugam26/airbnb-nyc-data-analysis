# Airbnb NYC Dataset - Data Cleaning

## Project Objective

This notebook applies the data cleaning plan developed during the Data Understanding & Assessment phase. Missing values, incorrect data types, invalid records, and data quality issues are addressed to prepare the dataset for exploratory data analysis.

## Cleaning Strategy

| Column | Issue                                    | Cleaning Strategy                                                                 | Reason                                                                                                                           |
| ------ | ---------------------------------------- | --------------------------------------------------------------------------------- | --------------------------------------------------------------------------------------------------------------------------- |
| `name` | 16 missing values, 4 numeric-only titles | Replaced missing values with **"No Listing Title"**; retained numeric-only titles | Preserved all records without introducing fabricated listing names. Numeric-only titles could not be confidently corrected. |
| `host_name` | 21 missing values, numeric-only names | Replaced missing values with **"No Host Name"**; retained numeric-only names | Preserved all records without introducing fabricated host names. Numeric-only values could not be reliably corrected. |
| `neighbourhood_group` | Stored as object | Converted to category | Contains a small, fixed set of repeated values. Using the category data type reduces memory usage and accurately represents the categorical nature of the data. |
| `neighbourhood` | Stored as object | Converted to category | Contains a finite set of repeated neighbourhood names. Converting to category reduces memory usage, improves performance, and accurately represents the categorical nature of the data. |
| `room_type` | Stored as object | Converted to category | Contains only three repeated categories. Converting to category reduces memory usage, improves performance, and accurately represents the categorical nature of the data. |
| `last_review`	| Stored as object; expected missing values	| Converted to datetime; retained NaT values |	Missing values correctly indicate listings with no reviews. Converting to datetime enables date-based analysis while preserving data integrity. |
| `price`	| 11 zero-priced listings	| Removed zero-priced listings; retained genuine high-priced listings	| Zero-priced listings likely represent invalid or missing pricing information. Since they account for only 0.02% of the dataset and the correct prices could not be recovered, removing them improves the reliability of price-based analyses while having a negligible impact on the dataset. |
| `minimum_nights` | Statistical upper outliers | Retained all values | The outliers represent legitimate long-term rental policies rather than data quality issues. No evidence was found to justify removing or modifying these records.

## 1. Cleaning the `name` Column

```python
df['name'] = df['name'].fillna('No Listing Title')
```

```python
df[df['name']=='No Listing Title'].shape[0]
```

```python
df['name'].isnull().sum()
```

## 2. Cleaning the `host_name` Column

```python
df['host_name'] = df['host_name'].fillna('No Host Name')
```

```python
df[df['host_name']=='No Host Name'].shape[0]
```

```python
df['host_name'].isnull().sum()
```

## 3. Cleaning the `neighbourhood_group` Column

```python
df['neighbourhood_group'] = df['neighbourhood_group'].astype('category')
```

```python
df['neighbourhood_group'].dtype
```

## 4. Cleaning the `neighbourhood` Column

```python
df['neighbourhood'] = df['neighbourhood'].astype('category')
```

```python
df['neighbourhood'].dtype
```

```python
df['neighbourhood'].nunique()
```

## 5. Cleaning the `room_type` Column

```python
df['room_type'] = df['room_type'].astype('category')
```

```python
df['room_type'].dtype
```

```python
df['room_type'].nunique()
```

## 6. Cleaning the `last_review` Column

```python
df['last_review'] = pd.to_datetime(df['last_review'])
```

```python
df['last_review'].dtype
```

## 7. Cleaning the `price` Column

```python
df.drop(df[df['price']==0].index, inplace=True)
```
